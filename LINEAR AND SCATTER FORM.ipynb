# MATPLOTLIB

```python
import numpy as np
```

```python
from matplotlib import pyplot as plt
```

```python
from matplotlib import style # FOR ADDING STYLE
```

# LINEAR GRAPH

```python
x=np.array([1,2,3])
y=np.array([2,3,5])
plt.plot(x,y,label="linear line",color="g")
plt.title("Simple linear graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()
```

# Scatter plot

```python
x=np.array([1,2,3])
y=np.array([2,3,5])
plt.scatter(x,y,label="Scatter Plot",color="g",marker="x")
plt.title("Simple scatter graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True,color="k")
plt.legend()
plt.show()
```
